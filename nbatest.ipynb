```python
#FUNCTIONS
def add_year(c_age):
    new_age = c_age+1
    return new_age
```

```python
dm['Age'].apply(add_year)
```

```python
dm[['FT', 'FTA']]

def ft_calculator(rw):
    ft_perc = rw['FT'] / (rw['FTA'] + 0.0001 )
    return round(ft_perc,2)
```

```python
dm.apply(ft_calculator, axis=1)
```

```python
ftt = dm.groupby('Player').sum().apply(ft_calculator, axis=1).sort_values(ascending = False)
ftt.head(100)
```
